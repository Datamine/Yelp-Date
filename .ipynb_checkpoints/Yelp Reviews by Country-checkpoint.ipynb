```python
import re
from matplotlib import pyplot
from collections import Counter
%matplotlib inline
import json
import pandas as pd
import numpy as np
```

```python
#nb germany reviews are in german
#quebecois are in french!!
```

```python
reviews = []
with open("dataset/yelp_academic_dataset_review.json") as data_file:
    for line in data_file:
        reviews.append(json.loads(line))

reviews_df = pd.DataFrame(reviews)
del reviews
```

```python
businesses = []
with open("dataset/yelp_academic_dataset_business.json") as data_file:
    for line in data_file:
        businesses.append(json.loads(line))
        
businesses_df = pd.DataFrame(businesses)
del businesses
```

```python
states = set(businesses_df['state'])
# One business in our dataset was mislabeled as being in Tamaulipas, Mexico.
states.remove("TAM")
```

```python
germany_states = ["BW", "NW"]
scotland_states = ["EDH", "SCB", "ELN", "FIF", "HAM", "KHL", "MLN", "NTH", "SCB", "XGL"]
canada_states = ["QC", "ON"]
us_states = [state for state in states if not ((state in germany_states) or 
                                               (state in scotland_states) or 
                                               (state in canada_states))]

location_mapping = dict([(x,"Germany") for x in germany_states] + 
                        [(x,"Scotland") for x in scotland_states] + 
                        [(x,"Canada") for x in canada_states] + 
                        [(x,"USA") for x in us_states])
```

```python
businesses_df['location'] = businesses_df['state'].map(location_mapping)
# this will leave one business with an empty location, since we dropped "TAM."
# remove that business and its associated reviews.
bad_biz_id = businesses_df[businesses_df['location'].isnull()]['business_id'].item()
businesses_df = businesses_df[businesses_df['location'].notnull()]
reviews_df = reviews_df[reviews_df['business_id'] != bad_biz_id]
```

```python
business_loc_dict = businesses_df.set_index('business_id').to_dict()['location']
reviews_df['location'] = reviews_df['business_id'].map(business_loc_dict)
```

```python
reviews = {u"Germany":Counter(), 
           u"Scotland":Counter(), 
           u"Canada":Counter(), 
           u"USA":Counter()}

for index, review in reviews_df.iterrows():
    biz_loc = review['location']
    reviews[biz_loc][review['stars']] += 1
```

```python
for place, counter in reviews.items():
    total_stars = float(sum(counter.values()))
    proportions = {k:v/total_stars for (k,v) in counter.items()}
    pyplot.bar(counter.keys(), [x*100 for x in proportions.values()])
    pyplot.xlabel("Star Rating")
    pyplot.axis([0,7,0,60])
    pyplot.ylabel("Percent of all Reviews ")
    pyplot.title(place)
    pyplot.show()
```

```python
# Considering the difference in review distribution, I wonder if Yelp does any kind of normalization
# on business ratings. If not, we would expect business ratings to be distributed similarly to the above.
# N.b. Yelp asseses business ratings down to the half-star level.
```

```python
business_ratings = {u"Germany":Counter(), 
                    u"Scotland":Counter(), 
                    u"Canada":Counter(), 
                    u"USA":Counter()}

for index, review in businesses_df.iterrows():
    biz_loc = review['location']
    business_ratings[biz_loc][review['stars']] += 1
```

```python
for place, counter in business_ratings.items():
    total_stars = float(sum(counter.values()))
    proportions = {k:v/total_stars for (k,v) in counter.items()}
    pyplot.bar(counter.keys(), [x*100 for x in proportions.values()], width=0.5)
    pyplot.xlabel("Star Rating")
    pyplot.axis([0,7,0,60])
    pyplot.ylabel("Percent of all Businesses")
    pyplot.title(place)
    pyplot.show()
```

```python
# the above suggests that the plethora of one-star reviews substantially evens out the five-star reviews in the US,
# resulting in a platykurtic distribution. it also, quite importantly, suggests that yelp does not normalize
# its reviews, but rather lets different geographies have different review distributions, which culturally
# is probably more useful for the residents.
```

```python
def cleanup(string):
    return re.sub('[^0-9 a-zA-Z]+', '', string)

reviews_df['text'] = reviews_df['text'].map(cleanup)
```

```python
review_lengths = {u"Germany":Counter(), 
                  u"Scotland":Counter(), 
                  u"Canada":Counter(), 
                  u"USA":Counter()}

for index, review in reviews_df.iterrows():
    biz_loc = review['location']
    review_lengths[biz_loc][len(review['text'])] += 1
```

```python
country_polyfits = {}

for place, counter in review_lengths.items():
    total_reviews = float(sum(counter.values()))
    proportions = {k:v/total_reviews for (k,v) in counter.items()}
    x_values = counter.keys()
    y_values = [x*100 for x in proportions.values()]
    
    # naming convention: "you've gotta be kidding me"
    x_new = np.linspace(x_values[0], x_values[-1], 2*len(x_values))
    coefficients = np.polynomial.polynomial.polyfit(x_values, y_values, 20)
    ffit = np.polynomial.polynomial.polyval(x_new, coefficients)
    pyplot.plot(x_new, ffit, 'c')
    
    country_polyfits[place] = coefficients
    
    # bar plot
    pyplot.bar(counter.keys(), y_values)
    pyplot.xlabel("Word Count")
    pyplot.ylabel("Percent of all Reviews")
    sample_size = "Sample Size = " + str(int(total_reviews)) + " Reviews" 
    pyplot.title(place)
    annotation_box = dict(boxstyle="square,pad=0.3", fc="white")
    pyplot.annotate(sample_size, xy=(2250, 0.10), bbox=annotation_box)
    pyplot.show()
```
